```python
try:
    input = raw_input   
except NameError:
    pass

from lark import Lark
from random import randint

def rand():
    return randint(0, 100)/100

# This function belongs to David Simmons-Duffin, Copyright 2010 
# Found on Quora webpage 
# https://www.quora.com/What-is-the-easiest-way-to-find-the-longest-common-prefix-or-suffix-of-two-sequences-in-Python
def commonprefix(m):
    "Given a list of pathnames, returns the longest common leading component"
    if not m: return ''
    s1 = min(m)
    s2 = max(m)
    for i, c in enumerate(s1):
        if c != s2[i]:
            return s1[:i]
    return s1

PRE_grammar = """       
    start: star+
         | exp+
         | fun
    
    fun: regex+ "=" TXT       -> odds
        | "generate" N regex+ -> generate
    
    regex: "(" regex ")"
         | star+
         | exp+
    
    star: exp "*" PROB     -> star
        | "(" star+ ")"
    
    exp: exp "|" PROB exp  -> or
        | "(" exp+ ")"
        | TXT              -> txt 
    
    TXT: LETTER+
    PROB: /[0][.][1-9][0-9]*|[0][.][0][1-9]+/
    N: /[1-9][0-9]*/
    
    %import common.LETTER
    %import common.WS
    %ignore WS
"""

# build parse-tree
parser = Lark(PRE_grammar)

# global variables for accumulating results
acc = 1 
txt_out = ""

# reset global variables
def reset():
    global acc 
    global odds
    global txt_out
    acc = 1 
    odds = 0
    txt_out = ""

# use the parse-tree to evaluate generic input
def run_instruction(t):
    global txt_out
    
    # use the parse-tree to generate random strings based on PRE
    def run_generate(t):
        global acc 
        global txt_out
                
        if t.data == 'or': 
            prob = float(t.children[1])
            if prob > rand():       
                run_generate(t.children[0])
                acc *= prob
            else:
                run_generate(t.children[2])            
                acc *= 1-prob 
        elif t.data == 'star':
            if(len(t.children)==1):
                run_generate(t.children[0])
            else:    
                prob = float(t.children[1])
                while(prob < rand()):
                    acc *= (1-prob)  
                    run_generate(t.children[0])
                acc *= prob 
        elif t.data == 'regex':
            for c in t.children: 
                run_generate(c)       
        elif t.data == 'exp':
            for c in t.children: 
                run_generate(c)
        elif t.data == 'txt':
            txt_out += t.children[0]
        else:
            raise SyntaxError('Unknown instruction: %s' % t.data)
    # end of run_generate
    
    # use the parse-tree to get the probability of a PRE producing af particular string
    def run_odds(t):
        global odds  # for the accumulated odds that PRE == TXT
        global txt_out
                
        if t.data == 'or': 
            pass
            #run through both children
        elif t.data == 'star':
            pass
            # price of N elements = exit-price * stay-price**N
        elif t.data == 'regex':
            for c in t.children: 
                run_odds(c)       
        elif t.data == 'exp':
            for c in t.children: 
                run_odds(c)
        elif t.data == 'txt':
            pass
            # use commonprefix(lst) to search for common prefix
        else:
            raise SyntaxError('Unknown instruction: %s' % t.data)
    # end of run_odds
    
    # ["generate" N PRE+]
    if t.data == 'generate':
        children = t.children    
        n = int(children[0])
        for c in children[1:]: 
            while(n > 0):
                run_generate(c)
                n -= 1  
                print("Probability of ", txt_out,"=", acc)
                reset()
    
    # [PRE+ "=" TXT]      
    elif t.data == 'odds':
        # PROPOSED ALGORITHM
        # traverse through the tree looking for a prefix match for TXT
        # then factor in the probability of this prefix. Delete prefix from TXT.
        # repeat process untill full match or untill the tree has been traversed fully
        # without any prefix match.
        txt_out = t.children[-1]
        for c in t.children[:-1]:
            run_odds(c)
        print("Probability of", txt_out,"is", odds)
        reset()
    else:
            raise SyntaxError('Unknown instruction: %s' % t.data)
# end of run_instruction

def run_PRE(program):
    parse_tree = parser.parse(program)
    for inst in parse_tree.children:
        run_instruction(inst)

def main():
    while True:
        code = input('> ')
        try:
            run_PRE(code)
        except Exception as e:
            print(e)

if __name__ == '__main__':
    main()
    
"""
TEST DATA:

THE GRAMMAR FOR STAR DOES NOT ACCEPT STAR
THUS IT CANNOT ACCIDENTLY USE STAR ON AN EMPTY STRING
(a*0.3)*0.4 -> ERROR

POSITIVE test cases:

generate 1 a b c
generate 3 a |0.4 b |0.3 c
generate 3 (a |0.3 b | 0.4 c)
generate 3 a*0.3 b*0.2
generate 3 (a|0.5b)*0.8 (c|0.3d)*0.5

NEGATIVE test cases:
SADLY, THESE DO NOT WORK

generate 3 a*0.3 (b*0.2) c
generate 3 a |0.3 b*0.4
generate 3 (a |0.2 b |0.3 c |0.4 d |0.5 r)* 0.2
"""
```
